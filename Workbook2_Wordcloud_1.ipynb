```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud,STOPWORDS

# 단어를 이 그림위에 겹쳐보이게 하기 위해 마스킹 데이터를 뽑는다.
alice_mask=np.array(Image.open
                    (r"C:\Users\ellka\Downloads\alice.png"))

# STOPWORDS(제외하는 단어)목록에 "said"를 추가한다.
# 빈도수를 계산할 때 다음 단어를 제외하고 계산한다.
stop = set(STOPWORDS)
stop.add("said")

# 배경색과 단어의 수,겹쳐보이게 할 마스크, stopwords를 지정한다.
wc = WordCloud(background_color="white",max_words=1000,
             mask=alice_mask,stopwords=stop)
WC = wc.generate(text)


WC.words_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))
plt.imshow(wc,interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw

# 텍스트의 이름이 constitution이다.
ko_doc = kolaw.open("constitution.txt").read()
```

```python
ko_doc
```

```python
from konlpy.tag import Okt
t = Okt()
ko_token = t.nouns(ko_doc)
ko_token
```

```python
kor = nltk.Text(ko_token,name = "대한민국 헌법")
print(kor)
print("총 단어(명사)의 수:",len(kor.tokens))
print("총 단어(명사)종류의 수:",len(set(kor.tokens)))
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

# 한글 폰트를 지정한다.
font_path = "C:\Windows\Fonts\H2GTRM.TTF"
font_name = fm.FontProperties(fname = font_path).get_name()

plt.rc('font',family = font_name)

plt.figure(figsize = (12,6))
kor.plot(50) # 빈도수 상위 50개만 출력한다.
plt.show()
```

```python
stop_words = ['제','정','관','그','이','바','위','및',
            '안','장','수','때']
kor = [each_word for each_word in kor 
       if each_word not in stop_words]

kor = nltk.Text(kor,name = "대한민국 헌법")
plt.figure(figsize = (12,6))
kor.plot(50) # 빈도수 상위 50개만 출력한다.
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 빈도수가 가장 높은 2단어 200개를 kor_data에 저장한다.
kor_data = kor.vocab().most_common(200)

# 한글 폰트를 지정한다.
font_path = "C:\Windows\Fonts\H2GTRM.TTF"

wc = WordCloud(font_path=font_path, background_color="white")

"""
generate_from_frequencies는 미리 정의된 단어의 빈도수를 이용해
워드 클라우드를 생성한다. kor_data에 저장된 단어와 그 빈도수를 
통해 워드 클라우드를 만든다.
"""
WC = wc.generate_from_frequencies(dict(kor_data))

plt.figure(figsize = (10,5))
plt.imshow(WC,interpolation="bilinear")
plt.axis("off")
plt.show()
```
