```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

# 한글 폰트를 지정한다.
font_path = "C:\Windows\Fonts\H2GTRM.TTF"
font_name = fm.FontProperties(fname = font_path).get_name()
plt.rc('font',family = font_name)

# stop_word를 지정한다.
stop_words = ['속보','대통령','오늘','조국','한국','첫','명','중'
              ,'전','문','첫','것','차','등','위','손','안','낮']

ko = [each_word for each_word in ko
    if each_word not in stop_words]
''' 
위의 한 줄은 다음 네 줄로 대체 가능하다.

for w in ko: 
    if w not in stop_words: 
        ko.append(w)
'''

ko = nltk.Text(ko, name = "워드 클라우드")

plt.figure(figsize = (12,6))
ko.plot(50) # 빈도수 상위 50개만 출력한다.
plt.show()
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image

# 빈도수가 가장 높은 단어 300개를 news에 저장한다.
news = ko.vocab().most_common(300)

# 한글 폰트를 지정한다.
font_path = "C:\Windows\Fonts\H2GTRM.TTF"

# 마스킹 데이터를 뽑는다.
korea_map=np.array(Image.open
                    (r"C:\Users\ellka\Downloads\korea_map.png"))

wc = WordCloud(font_path=font_path, background_color="white",
                mask=korea_map)
WC = wc.generate_from_frequencies(dict(news))

plt.figure(figsize = (12,8))
plt.imshow(WC,interpolation="bilinear")
plt.axis("off")
plt.show()
```
