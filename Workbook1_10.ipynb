```python
Seoul_pop = pd.read_excel('Seoul_population.xls', 
                          header = 2,
                          usecols = 'B, D, G, J, N',
                          encoding='utf-8')
Seoul_pop.head()
```

```python
Seoul_pop.rename(columns={Seoul_pop.columns[0] : '구별', 
                          Seoul_pop.columns[1] : '인구수', 
                          Seoul_pop.columns[2] : '한국인', 
                          Seoul_pop.columns[3] : '외국인', 
                          Seoul_pop.columns[4] : '고령자'}, inplace=True)
Seoul_pop.head()
```

```python
Seoul_CCTV.sort_values(by='소계', ascending=True).head(5)
```

```python
Seoul_CCTV.sort_values(by='소계', ascending=False).head(5)
```

```python
Seoul_CCTV['최근증가율'] = (Seoul_CCTV['2018년'] + Seoul_CCTV['2017년'] + \
                        Seoul_CCTV['2016년'] + Seoul_CCTV['2015년']) / \
                        (Seoul_CCTV['2014년'] + Seoul_CCTV['2013년'] + \
                        Seoul_CCTV['2012년'] + Seoul_CCTV['2011년 이전']) * 100
Seoul_CCTV.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
Seoul_pop.head()
```

```python
Seoul_pop.drop([0], inplace=True)
Seoul_pop.head()
```

```python
Seoul_pop['구별'].unique()
```

```python
Seoul_pop[Seoul_pop['구별'].isnull()]
```

```python
Seoul_pop.drop([26], inplace=True)
Seoul_pop.head()
```

```python
Seoul_pop['외국인비율'] = Seoul_pop['외국인'] / Seoul_pop['인구수'] * 100
Seoul_pop['고령자비율'] = Seoul_pop['고령자'] / Seoul_pop['인구수'] * 100
Seoul_pop.head()
```

```python
data_result = pd.merge(Seoul_CCTV, Seoul_pop, on='구별')
data_result.head()
```

```python
del data_result['2011년 이전']
del data_result['2012년']
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
plt.figure()
data_result['소계'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result['소계'].sort_values().plot(kind='barh', 
                                     grid=True, figsize=(10,10))
plt.show()
```

```python
data_result['CCTV비율'] = data_result['소계'] / data_result['인구수'] * 100

data_result['CCTV비율'].sort_values().plot(kind='barh', 
                                         grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
import numpy as np

fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(23):
    plt.text(data_result['인구수'][n], data_result['소계'][n], 
             data_result.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```
